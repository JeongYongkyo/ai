```python
def village(L):
    if len(L) % 2 == 0:
        mid = (len(L)//2)-1
        avg = int((L[mid]+L[mid+1])/2)
    else :
        mid = len(L)//2
        avg = int(L[mid])
    
    Sum = 0
    for i in L:
        Sum += abs(i-avg)
    return Sum
    
t = int(input())
for _ in range(t):
    L = list(map(int,input().split()))
    print(village(L))
```

```python
#TIME LIMIT
def village(L):
    min = sum(L)
    for i in range(max(L)+1):
        Sum = 0
        for j in L:
            Sum += abs(i-j)
        if min > Sum:
            min = Sum
    return min

t = int(input())
for _ in range(t):
    L = list(map(int,input().split()))
    print(village(L))
```
