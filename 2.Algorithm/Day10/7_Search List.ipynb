```python
t = int(input())
for _ in range(t):
    L1 = list(map(int,input().split()))
    L2 = list(map(int,input().split()))
    for a in L2:
        if a in L1:
            print("YES", end=" ")
        else:
            print("NO", end=" ")
    print("")
```
