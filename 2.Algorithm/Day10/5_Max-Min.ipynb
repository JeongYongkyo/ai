```python
def Max(L):
    m = L[0]
    for i in L:
        if m < i:
            m = i
    return m

def Min(L):
    m = L[0]
    for i in L:
        if m > i:
            m = i
    return m

T = int(input())
for _ in range(T):
    L = list(map(int,input().split()))
    print(Max(L) - Min(L))
```
