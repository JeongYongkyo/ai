```python
t = int(input())
for _ in range(t):
    N = int(input())
    L = list(map(int, input().split()))
    
    lap = []
    rank = {}
    stop = False
    for i in L:
        if i in rank:
            for idx in range(rank[i]-1, -1, -1):
                if lap[idx] == lap[rank[i]]:
                    print("YES")
                    stop = True
                    break
                else:
                    break
            if stop:
                break
            else:
                lap[rank[i]] += 1
        else:
            rank[i] = len(lap)
            lap.append(1)
    if stop:
        continue
    else:
        print("NO")
```
