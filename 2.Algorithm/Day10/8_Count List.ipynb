```python
t = int(input())
for _ in range(t):
    L = list(map(int,input().split()))
    C = list(map(int,input().split()))
    for i in C:
        print(L.count(i), end=" ")
    print("")
```

```python
def Count(x,l):
    cnt = 0
    for i in l:
        if x == i:
            cnt += 1
    return cnt

t = int(input())
for _ in range(t):
    l = list(map(int,input().split()))
    qlist = list(map(int,input().split()))
    res = []
    for q in qlist:
        res.append(l.Count(q))
    print(*res)
```
