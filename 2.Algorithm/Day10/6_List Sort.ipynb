```python
n = int(input())
for i in range(n):
    L = list(map(int, input().split()))
    L.sort(reverse=True)
    print(*L)
```

```python
def Qsort_rev(l):
    print(type(l))
    n = len(l)
    if n <= 1:
        return 1
    pivot = l[0]
    high = []
    middle = [pivot]
    low = []
    for i in range(1,n):
        if l[i] > pivot:
            high.append(l[i])
        elif l[i] < pivot:
            low.append(l[i])
        else:
            middle.append(l[i])
    return Qsort_rev(high) + middle + Qsort_rev(low)

n = int(input())
for _ in range(n):
    L = list(map(int, input().split()))
    
Qsort_rev(L)
print(*L)
```
