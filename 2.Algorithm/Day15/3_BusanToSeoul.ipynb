```python
def GoToSeoul():
    if G > L : return 0
    
    distGap = []
    for x in range(len(GasStation)+1):
        if x == 0: distGap.append(GasStation[0])
        elif x < len(GasStation):
            distGap.append(GasStation[x]-GasStation[x-1])
        else:
            distGap.append(L-GasStation[x-1])
            break

    myLocation = 0; ans = 0
    for y in distGap:
        if y > G: return -1
        if myLocation + y <= G:
            myLocation += y
        else:
            ans += 1
            myLocation = y
    return ans
            
T = int(input())
for _ in range(T):
    G,L = map(int,input().split())
    GasStation = list(map(int,input().split()))
    
    print(GoToSeoul())
```

```python
# 조교님

t = int(input())
for _ in range(t):
    G, l = map(int,input().split())
    station = list(map(int,input().split())) + [l] #맨 마지막 목적지
    present = 0
    rescount = 0
    
    for i in range(len(station)-1):
        if station[i] - present <= G: # 만약 present에서 이번 주유소까지 갈 수 있다면
            if station[i+1] - present <= G:
                continue
            else:
                present = station[i]
                rescount += 1
        else:
            rescount = -1
            break
    
    if l - present <= G:
        print(rescount)
    else:
        print(-1)
```
