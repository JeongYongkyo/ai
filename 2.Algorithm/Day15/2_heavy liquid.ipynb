```python
t = int(input())
for _ in range(t):
    N,C = map(int,input().split())
    M = []
    for _ in range(N):
        W,V = map(int,input().split())
        M.append((W/V,W,V))
    M.sort(reverse=True)
    print(M)
    ans = 0
    for a,b,c in M:
        if C >= c:
            ans += b
            C -= c
        else:
            ans += C*a
            break
    print(int(ans))
```

```python
# 조교님
def cmp(x):
    return x[0]/x[1]

t = int(input())
for _ in range(t):
    N,C = map(int,input().split())
    liquidlist = []
    for i in range(N):
        liquidlist.append(tuple(map(int,input().split())))
    liquidlist.sort(key=cmp,reverse=True)
    
    maxg = 0
    print(liquidlist)
    for i in range(N):
        if C >= liquidlist[i][1]:
            maxg += liquidlist[i][0]
            C -= liquidlist[i][1]
        else:
            maxg += C*(liquidlist[i][0]/liquidlist[i][1])
            break
    
    print(int(maxg))
```
