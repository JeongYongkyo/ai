```python
# 조교님
t = int(input())
coin = [50000,10000,5000,1000,500,100]
for _ in range(t):
    n = int(input())
    coinnum = 0
    for i in coin:
        coinnum += n // i
        n = n % i
        if n == 0:
            break
    print(coinnum)
```

```python
def calc(tax):
    won_50000 = tax // 50000
    won_10000 = (tax % 50000) // 10000    
    won_5000 = ((tax % 50000) % 10000) // 5000    
    won_1000 = (((tax % 50000) % 10000) % 5000) // 1000    
    won_500 = ((((tax % 50000) % 10000) % 5000) % 1000) // 500    
    won_100 = (((((tax % 50000) % 10000) % 5000) % 1000) % 500)//100
    
    return won_50000+won_10000+won_5000+won_1000+won_500+won_100

T = int(input())
for _ in range(T):
    tax = int(input())
    print(calc(tax))
```
