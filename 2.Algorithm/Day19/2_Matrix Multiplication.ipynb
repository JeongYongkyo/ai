```python
# for (1 .............x................n)
# for (0..............i...............n-x)
# for (i..............k...............i+x)
```

```python
t = int(input())
for _ in range(t):
    n = int(input())
    L = [list(map(int, input().split())) for _ in range(n)]
    D = [[0] * n for _ in range(n)]
    for x in range(1, n):
        for i in range(n - x):
            j = i + x
            print("x,i,j :",x,i,j)
            D[i][j] = float("inf")
            for k in range(i, j):
                print("i,j,k :",i,j,k)
                print(L[0])
                print(L[1])
                print(L[2])
                print(L[3])
                print("-----------------")
                print(D[0])
                print(D[1])
                print(D[2])
                print(D[3])
                print("")
                D[i][j] = min(D[i][j], D[i][k] + D[k + 1][j] + L[i][0] * L[k][1] * L[j][1])
    print(D[0][n - 1])
```
