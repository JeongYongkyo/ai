```python
t = int(input())
#t = 1
for _ in range(t):
    n,C = map(int,input().split())
    W = list(map(int,input().split()))
    V = list(map(int,input().split()))
    S = [[0]*(C+1) for _ in range(n)]
    for x in range(C+1):
        if x >= W[0]:
            #print("start :",V[0])
            S[0][x] = V[0]
            
    for i in range(1,n):
        for j in range(C+1):
            #print("i,j :",i,j)
            if W[i] > j:
                #print("S[i-1][j] :",S[i-1][j])
                S[i][j] = S[i-1][j]
            else:
                #print("max(S[i-1][j],S[i-1][j-W[i]]+V[i]) :",S[i-1][j],S[i-1][j-W[i]],V[i])
                S[i][j] = max(S[i-1][j],S[i-1][j-W[i]]+V[i])
    #print("S :",S[0])
    #print("S :",S[1])
    #print("S :",S[2])
    #print("S :",S[3])
    print(S[n-1][C])
```
