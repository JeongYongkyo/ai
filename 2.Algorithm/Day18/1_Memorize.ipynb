```python
t = int(input())
for _ in range(t):
    n,m = map(int,input().split())
    M = [[0]*m for _ in range(n)]
    S = [[0]*m for _ in range(n)]
    for i in range(n):
        M[i] = list(map(int,input().split()))
    #print("M :",M)
    S[0][0] = M[0][0]

    for x in range(1,m):
        S[0][x] = S[0][x-1]+M[0][x]
    
    for y in range(1,n):
        S[y][0] = S[y-1][0]+M[y][0]
        
    for x in range(1,m):
        for y in range(1,n):
            #print("x, y :",x,y)
            S[y][x] = max(S[y][x-1],S[y-1][x])+M[y][x]
            #print("down :",S[y][x-1])
            #print("right :",S[y-1][x])
            #print("M[y][x] :",M[y][x])
    print(S[n-1][m-1])
```

```python
# 조교님
t = int(input())
for _ in range(t):
    n,m = map(int,input().split())
    data = []
    for i in range(n):
        data.append(list(map(int,input().split())))
    T = [[0]*m for i in range(n)]
    for i in range(n):
        for j in range(m):
            if i == 0 and j == 0:
                T[i][j] = data[i][j]
            elif i ==0:
                T[i][j] = T[i][j-1] + data[i][j]
            elif j ==0:
                T[i][j] = T[i-1][j] + data[i][j]
            else:
                T[i][j] = max(T[i][j-1],T[i-1][j])+data[i][j]
    print(T[n-1][m-1])
```
