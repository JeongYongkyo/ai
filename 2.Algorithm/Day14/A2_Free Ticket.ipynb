```python
# TIMELIMIT
import heapq

def function(graph):
    pq = [[0,0]]
    dist = [-1]*V
    visited = [False]*V
    while pq:
        d,u = heapq.heappop(pq)
        if visited[u]:
            continue
        visited[u] = True
        dist[u] = d
        for x in graph[u]:
            if x:
                if not visited[x[0]]:
                    heapq.heappush(pq,(x[1]+d,x[0]))
    return(dist[-1])

t = int(input())
for _ in range(t):
    V, E = map(int,input().split())
    graph = [[] for _ in range(V)]

    for _ in range(E):
        u, v, c = map(int,input().split())
        graph[u].append([v, c])
    ans = []
    for x in range(len(graph)) :
        for y in range(len(graph[x])):
            if graph[x][y]:
                tmp = graph[x][y][1]
                graph[x][y][1] = 0
            ans.append(function(graph))
            graph[x][y][1] = tmp
    if ans:
        print(min(ans))
    else:
        print(-1)
```
