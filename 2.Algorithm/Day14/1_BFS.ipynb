```python
from collections import deque

t = int(input())
for _ in range(t):
    N, M = map(int,input().split())
    List = [[] for _ in range(N)]
    
    for _ in range(M):
        u, v = map(int,input().split())
        List[u].append(v)
    
    for i in range(N):
        List[i].sort(reverse=False)
        
    visited = [False]*N
    queue=deque([0])
    ans=[]

    while queue:
        v = queue.popleft()
        if not visited[v]:
            visited[v] = True
            ans.append(v)
            for i in List[v]:
                if not visited[i]:
                    queue.append(i)    
    print(*ans)
```
