```python
# 조교님 
import heapq

t = int(input())
for _ in range(t):
    (V, E) = map(int,input().split())
    el = [[] for _ in range(V)]
    
    for _ in range(E):
        (u,v,c) = map(int,input().split())
        el[u].append((v, c))
    
    pq=[(0,0)]
    dist=[-1]*V
    visited=[False]*V
    while pq:
        (d,u)=heapq.heappop(pq)
        if visited[u]:
            continue
        visited[u] = True
        dist[u]=d
        for (v,c) in el[u]:
            if not visited[v]:
                heapq.heappush(pq,(c+d,v))
    print(dist[-1])
        
```

```python
# 조교님 인접행렬 + 배열
INF = float('inf')
T = int(input())
for _ in range(T):
    N,M = map(int,input().split())
    Matrix = [[0]*N for _ in range(N)]
    for i in range(M):
        u,v,c = map(int,input().split())
        Matrix[u][v] = c
    D = [0]+[INF]*(N-1)
    Check = [False]*N
    for iteration in range(N):
        Min = INF
        for i in range(N):
            if Check[i] == False and D[i] < Min:
                Min = D[i]
                idx = i
        if Min == INF : break
        Check[idx] = True
        for i in range(N):
            if Matrix[idx][i] != 0 and D[idx] + Matrix[idx][i] < D[i]:
                D[i] = D[idx] + Matrix[idx][i]
    if Check[N-1] == False:
        print(-1)
    else:
        print(D[N-1])
```
