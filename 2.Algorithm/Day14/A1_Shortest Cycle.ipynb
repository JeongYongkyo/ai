```python
def DFS(graph, visited, u, v, k):
    Stack = [u]
    cnt = 0
    print(graph, visited, u, v, k)
    
    if u == v:
        if k > 2:
            return [k]
        else:
            return []
    
    while Stack:
        x = Stack.pop()
        if visited[x] == 1:
            continue
        cnt += 1
        print("x :",x)
        visited[x] = 1
        
        for i in graph[x]:
            #if i == x:
            #    continue
            if visited[i] == 0:
                Stack.append(i)
                k += 1
            print("Stack :",Stack)
            print("visit :",visited)
    
    print(cnt)
    print("")
    return cnt
    
#T = int(input())
T = 1
for _ in range(T):
    #V, E = map(int,input().split())
    V, E = [3, 3]
    graph = [[] for _ in range(V)]
    '''
    for _ in range(E):
        start, end = map(int,input().split())
        graph[start].append(end)
        graph[end].append(start)
    '''
    graph=[[1, 2], [0, 2], [1, 0]]
    ans = []
    for v in range(V):
        visited = [0]*V
        print("****************")
        print("v: ", v)
        if len(graph[v]) == 1 or visited[v]:
            continue
        else:
            for u in graph[v]:
                ans.append(DFS(graph, visited, u, v, 1))
            visited[v] = 1
    
    if ans:
        print(min(ans))
    else:
        print(-1)
```

```python
def explore(_edges, _passed, _u, _v, _k):
    print("u: ", _u, "v: ", _v, "k: ", _k)
    print(_passed)
    if _u == _v:
        if _k > 2:
            return [_k]
        else:
            return []

    r = []
    if len(_edges[_u]) == 1:
        _passed[_u] = 1
    else:
        for i in _edges[_u]:
            if _passed[i]:
                pass
            else:
                _passed[_u] = 1
                r += explore(_edges, _passed, i, _v, _k + 1)
                _passed[_u] = 0
        _passed[_u] = 1

    return r


T = int(input())
for t in range(T):
    V, E = [int(i) for i in input().split()]
    edges = {}
    passed = {}
    for v in range(V):
        edges[v] = []
        passed[v] = 0
    for e in range(E):
        v = [int(i) for i in input().split()]
        edges[v[0]].append(v[1])
        edges[v[1]].append(v[0])

    result = []
    print("")
    
    
    for v in range(V):
        print("********")
        print("v: ", v)
        print("edges: ", edges)
        print("passed: ", passed)
        if len(edges[v]) == 1 or passed[v]:
            pass
        else:
            for u in edges[v]:
                result += explore(edges, passed, u, v, 1)
        passed[v] = 1
        print("result: ", result)

    if result:
        print(min(result))
    else:
        print(-1)
```
