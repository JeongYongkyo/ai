```python
# 조교님
from collections import deque
T = int(input())
for _ in range(T):
    N,M = map(int,input().split())
    List = [[] for _ in range(N)]
    Indegree = [0]*N
    for _ in range(M):
        a, b = map(int,input().split())
        List[a].append(b)
        Indegree[b] += 1
    Queue = deque([])
    for i in range(N):
        if Indegree[i] == 0:
            Queue.append(i)
            break    
    result = []
    while Queue:
        x = Queue.popleft()
        result.append(x)
        for i in List[x]:
            Indegree[i] -= 1
            if Indegree[i] == 0:
                Queue.append(i)
    if len(result) == N:
        print(*result)
    else:
        print(-1)
```
