```python
def Pow(n, k ,m):
    if k == 0: return 1
    elif k == 1: return n % m
    elif k % 2 == 1: return (Pow(n,k//2,m)**2*n)%m
    elif k % 2 == 0: return (Pow(n,k//2,m)**2)%m

t = int(input())
for _ in range(t):
    n, k, m = map(int,input().split())
    print(Pow(n, k, m))
```

```python
# O(k) -> O(log2k)

def Power(n, k ,mod):
    if k == 1: return n % mod
    if k == 0: return 1
    
    tmp=Power(n,k//2,mod)
    if k % 2 == 0: return (tmp**2)%mod
    else: return (n*tmp**2)%mod

t = int(input())
for _ in range(t):
    n, k, mod = map(int,input().split())
    print(Power(n, k, mod))
```
