```python
def dist(a,b):
    return abs(a[0]-b[0])+abs(a[1]-b[1])

def divide(start, end):
    if start == end: return float("inf")
    elif end - start == 1: return dist(location[end],location[start])
    mid = (start+end) // 2
    center = min(divide(start,mid),divide(mid,end))
    
    candidate=[]
    for i in range(start,end+1):
        if abs(location[mid][0] - location[i][0]) < center:
            candidate.append(location[i])
    candidate.sort(key=lambda x:x[1])
    lc=len(candidate)
    for i in range(lc-1):
        for j in range(i+1,lc):
            dy = abs(candidate[i][1] - candidate[j][1])
            if dy < center:
                center = min(center,dist(candidate[i],candidate[j]))
            else:
                break
    return center

t = int(input())
for _ in range(t):
    N = int(input())
    location = []
    for _ in range(N):
        inputAxis = tuple(map(int,input().split()))
        location.append(inputAxis)
    location.sort()
    
    print(divide(0, len(location) - 1))
```
