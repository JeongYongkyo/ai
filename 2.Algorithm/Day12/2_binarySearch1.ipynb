```python
# 조교님 - 반복문
def binary(l,x):
    start = 0
    end = len(l)
    while start < end:
        mid=(start+end)//2
        if l[mid] == x: return mid
        if l[mid] > x: end=mid
        if l[mid] < x: start=mid+1
    return -1

t = int(input())
for _ in range(t):
    l = list(map(int,input().split()))
    q = list(map(int,input().split()))
    #ans=list(map(lambda x: binary(l,x),q))
    ans=[binary(l,x) for x in q]
    print(*ans)
```

```python
# 조교님 - 재귀
def binary(l, s, e, x):
    if s >= e: return -1
    mid = (s + e)// 2
    if l[mid] == x: return mid
    elif l[mid] > x: return binary(l, s, mid, x)
    else: return binary(l, mid+1, e, x)
    
t = int(input())
for _ in range(t):
    l = list(map(int,input().split()))
    q = list(map(int,input().split()))
    ans=[binary(l,0,len(l),x) for x in q]
    print(*ans)
```

```python
# 처음 생각한 거
def binarySearch(n,m):
    start = 0
    end = len(n)-1
    cnt = start+end//2
    result = []
    
    for i in m:
        if n[cnt] > i:    # 16 < 21
            start = cnt+1
        elif n[cnt] < i:
            end = cnt-1    # 4
        elif n[cnt] == i:
            result.append(i)
        else:
            result.append(-1)
    return result

t = int(input())
for _ in range(t):
    n = list(map(int,input().split()))
    m = list(map(int,input().split()))
    
    print(binarySearch(n,m))
```

```python
#연습
def binarySearch(n,q):
    start = 0
    end = len(n)
    while start < end:
        mid = (start + end) // 2
        if n[mid] == q: return mid
        elif n[mid] > q: end = mid
        elif n[mid] < q: start = mid+1
    return -1
    
t = int(input())
for _ in range(t):
    n = list(map(int,input().split()))
    m = list(map(int,input().split()))
    
    ans = [binarySearch(n,q) for q in m]
    print(*ans)
```
