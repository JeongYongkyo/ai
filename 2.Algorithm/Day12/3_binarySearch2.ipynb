```python
#조교님 - 반복문
def binary(l,x):
    if x <= l[0]: return l[0]
    if x >= l[-1]: return l[-1]
    start = 0
    end = len(l)
    while start < end:
        mid=(start+end)//2
        if l[mid] == x: return l[mid]
        elif x < l[mid]: end=mid
        else: start = mid+1
    if abs(l[end-1]-x) <= abs(l[end]-x): return l[end-1]
    else: return l[end]

t = int(input())
for _ in range(t):
    l = list(map(int,input().split()))
    q = list(map(int,input().split()))
    ans=[binary(l,x) for x in q]
    print(*ans)
```

```python
#조교님 - 재귀
def binary(l, s, e, x):
    if x < l[0]: return l[0]
    if x > l[-1]: return l[-1]
    if s >= e:
        if abs(l[e-1]-x) <= abs(l[e]-x): return l[e-1]
        else: return l[e]
    mid = (s + e)//2
    if l[mid] == x: return l[mid]
    elif l[mid] > x: return binary(l, s, mid, x)
    else: return binary(l, mid+1, e, x)
    
t = int(input())
for _ in range(t):
    l = list(map(int,input().split()))
    q = list(map(int,input().split()))
    ans=[binary(l,0,len(l),x) for x in q]
    print(*ans)
```

```python
# 연습
def binary(n, x):
    if x <= n[0]: return n[0]
    if x >= n[-1]: return n[-1]
    start = 0
    end = len(n)
    while start < end:
        mid=(start+end)//2
        if n[mid] == x: return x
        elif n[mid] > x: end = mid
        else:start = mid+1
    if abs(n[end-1]-x) <= abs(n[end]-x): return n[end-1]
    else: return n[end]
    
t = int(input())
for _ in range(t):
    n = list(map(int,input().split()))
    m = list(map(int,input().split()))
    ans = [binary(n, x) for x in m]
    print(*ans)
```
