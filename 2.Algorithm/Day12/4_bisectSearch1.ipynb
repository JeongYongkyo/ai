```python
# 조교님 - import bisect
import bisect

def find(l,x):
    i = bisect.bisect_left(l,x)
    if i != len(l) and l[i] == x:
        return i
    return -1

t = int(input())
for _ in range(t):
    l = list(map(int,input().split()))
    q = list(map(int,input().split()))
    ans=[find(l,x) for x in q]
    print(*ans) 
```
