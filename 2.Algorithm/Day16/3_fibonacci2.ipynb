```python
def Fibo(n):
    if n <= 2:
        return 1
    x, y = 0, 1
    for _ in range(2, n+1):
        S = x + y
        x = y
        y = S
    return S
        
T = int(input())

for _ in range(T):
    n = int(input())
    print(Fibo(n))
```

```python
# 조교님
T = int(input())
for _ in range(T):
    n = int(input())
    fibo = [0]*(n+1)
    for i in range(1,n+1):
        if i == 1 or i == 2:
            fibo[i] = 1
        else:
            fibo[i] = fibo[i-1] + fibo[i-2]
    print(fibo[n])
```
