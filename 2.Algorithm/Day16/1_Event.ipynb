```python
# 조교님
def cmp(x):
    return x[1]

T = int(input())
for _ in range(T):
    n = int(input())
    event = []
    for _ in range(n):
        event.append(tuple(map(int,input().split())))
    event.sort(key=cmp)
    #event.sort(key = lambda x:x[1], reverse=False)

    count78 = 0
    time = -1
    for i in range(n):
        if time < event[i][0]:
            count78 += 1
            time = event[i][1]
    print(count78)
```
