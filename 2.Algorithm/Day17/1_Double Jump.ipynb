```python
T = int(input())
for _ in range(T):
    n = int(input())
    data = list(map(int,input().split()))
    f = [0]*(n+1)
    
    for x in range(1,n+1):
        if x == 1: 
            f[x] = data[x-1]
        if x % 2 == 0: 
            f[x] = max(f[x-1],f[x-2],f[x//2])+data[x-1]
        else: 
            f[x] = max(f[x-1],f[x-2])+data[x-1]
    print(f[-1])
```

```python
# 조교님
def dbro(l,n):
    DP = [0]*len(l)
    for i in range(1,n+1):
        if i == 1:
            DP[i] = l[i]
        elif i % 2 == 1:
            pr = DP[i-1] + l[i]
            ju = DP[i-2] + l[i]
            DP[i] = max(pr,ju)
        else:
            pr = DP[i-1] + l[i]
            ju = DP[i-2] + l[i]
            do = DP[i//2] + l[i]
            DP[i] = max(pr,ju, do)
    return DP[n]

testnum = int(input())
for _ in range(testnum):
    n = int(input())
    lis = list(map(int,input().split()))
    lis = [0] + lis
    print(dbro(lis,n))
```
