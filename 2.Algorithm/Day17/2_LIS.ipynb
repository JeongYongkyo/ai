```python
T = int(input())
for _ in range(T):
    n = list(map(int,input().split()))
    dp = [1 for i in range(len(n))]
    for i in range(len(n)):
        for j in range(i):
            if n[i] > n[j]:
                dp[i] = max(dp[i], dp[j]+1)
                print(dp)
    print(max(dp))
```

```python
# 조교님
t = int(input())
for _ in range(t):
    data = list(map(int,input().split()))
    N = len(data)
    T = [0]*N
    for k in range(N):
        for j in range(k):
            if data[k] > data[j]:
                T[k] = max(T[k], T[j])
        T[k] += 1
        print(T)
    print(max(T))
```
