```python
# TIMELIMIT
import time
start_time = time.time()

def tile(n):
    dp = []
    mySum = [1, 3 , 10, 32]
    for i in range(n):
        if ( i == 0):
            dp.append(1)
        elif ( i == 1 ):
            dp.append(2)
        elif ( i == 2):
            dp.append(7)
        elif  ( i== 3 ):
            dp.append(22)
        else:
            tmp = 2*dp[i-1]
            tmp += 3*dp[i-2]
            tmp += 2*mySum[-3]
            
            mySum.append( (mySum[-1] + tmp)%1000000000 )
            dp.append(tmp%1000000000)
            
    return dp

#T = int(input())
T = 1
for _ in range(T):
    #n = int(input())
    n = 999999
    print(tile(n+1)[-1])
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
import time
start_time = time.time()
leng = 999999

dp1 = [0]*(leng+1)
dp1[0:2] = [1, 2]

dp2 = [0]*(leng+1)
dp2[0:2] = [1, 3]

for i in range(2, leng+1):
    dp1[i] = 2*dp2[i-1]
    dp1[i] %= 1000000000
    dp1[i] += dp1[i-2]
    dp1[i] %= 1000000000    
    
    dp2[i] = dp2[i-1] + dp1[i]
    dp2[i] %= 1000000000
    
print(dp1[leng])
print("--- %s seconds ---" % (time.time() - start_time))
```
