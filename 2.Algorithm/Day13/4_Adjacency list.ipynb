```python
t = int(input())
for _ in range(t):
    N, M = map(int,input().split())
    Adj_list = [[] for _ in range(N)]
    
    for _ in range(M):
        a, b = map(int,input().split())
        Adj_list[a].append(b)
        Adj_list[b].append(a)
       
    for i in range(N):
        Adj_list[i].sort()
        print(*Adj_list[i])
```

```python
# 조교님
t = int(input())
for _ in range(t):
    N, M = map(int,input().split())
    List = [[] for _ in range(N)]
    
    for _ in range(M):
        u, v = map(int,input().split())
        List[u].append(v)
        List[v].append(u)
        
    for i in range(N):
        List[i].sort()
        print(*List[i])
```
