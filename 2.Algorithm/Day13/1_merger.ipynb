```python
def mergeSort(n,m):
    n_S = m_S = 0
    ans = list()
    while True:
        if m_S == len(m) or n_S == len(n):
            break
        elif n[n_S] > m[m_S]: ans.append(2); m_S+=1
        elif n[n_S] < m[m_S]: ans.append(1); n_S+=1
        else: return False
    
    if m_S == len(m):
        for _ in range(len(n)-n_S):
            ans.append(1)
    else:
        for _ in range(len(m)-m_S):
            ans.append(2)
    return ans
    
t = int(input())

for _ in range(t):
    n = list(map(int,input().split()))
    m = list(map(int,input().split()))
    
    ans = mergeSort(n,m)
    print(*ans)
```

```python
# 조교님
def merge(l1, l2):
    n = len(l1)
    m = len(l2)
    i = j = 0
    ret=[0]*(n+m)
    for k in range(n+m):
        if j>=m:
            ret[k]=1
            i+=1
        elif i >= n:
            ret[k]=2
            j+=1
        elif l1[i]<l2[j]:
            ret[k]=1
            i+=1
        else:
            ret[k]=2
            j+=1
    return ret

t = int(input())
for _ in range(t):
    n = list(map(int,input().split()))
    m = list(map(int,input().split()))
    
    print(merge(n,m))
```
