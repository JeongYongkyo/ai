```python
# 조교님
T = int(input())
for _ in range(T):
    N, M = map(int,input().split())
    List = [[] for _ in range(N)]
    for _ in range(M):
        u, v = map(int,input().split())
        List[u].append(v)
        List[v].append(u)
    
    for i in range(N):
        List[i].sort(reverse=True)
    
    Check = [False]*N
    Stack = [0]
    while Stack:
        v = Stack.pop()
        if Check[v] == True:
            continue
        Check[v] = True
        print(v, end=" ")
        for i in List[v]:
            if Check[i] == False:
                Stack.append(i)
    print("")
```

```python
import sys
sys.setrecursionlimit(1000000)

def dfs(graph, start_node):
    visit = list()
    stack = list()

    stack.append(start_node)

    while stack:
        node = stack.pop()
        if node not in visit:
            visit.append(node)
            stack.extend(graph[node])
    
    return visit

t = int(input())
for _ in range(t):
    N, M = map(int,input().split())
    graph = dict()
    List = [[] for _ in range(N)]
    
    for _ in range(M):
        u, v = map(int,input().split())
        List[u].append(v)
        List[v].append(u)
    
    for i in range(N):
        List[i].sort(reverse=True)
        graph[i] = List[i]
        
    print(*dfs(graph,0))
```

```python
# 조교님
def DFS(v, List, Check):
    print(v, end=" ")
    Check[v] = True
    for i in List[v]:
        if Check[i] == False:
            DFS(i, List, Check)
    
t = int(input())
for _ in range(t):
    N, M = map(int,input().split())
    List = [[] for _ in range(N)]
    for _ in range(M):
        u, v = map(int,input().split())
        List[u].append(v)
        List[v].append(u)
    
    for i in range(N):
        List[i].sort()
    
    Check = [False]*N
    DFS(0,List,Check)
    print("")
```
