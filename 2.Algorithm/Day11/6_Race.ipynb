```python
from collections import deque

def raceRank(carNum):
    deq = deque([])
    for q in carNum:
        if q not in deq:
            deq.append(q)
        elif q in deq:
            if deq.count == 0:
                return("YES")
            else:
                if deq.popleft() != q:
                    return("YES")
    return("NO")

t = int(input())

for _ in range(t):
    L = list(map(int, input().split()))
    print(raceRank(L))
```

```python
from collections import deque

t = int(input())

for _ in range(t):
    car = list(map(int, input().split()))
    q = deque([])
    for c in car:
        if not q or q[0] != c:
            q.append(c)
        else:
            q.popleft()
    if q:
        print("YES")
    else:
        print("NO")
```
