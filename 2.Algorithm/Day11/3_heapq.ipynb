```python
import heapq

t = int(input())

for _ in range(t):
    hq = []
    n = int(input())
    for _ in range(n):
        m = int(input())
        if m > 0:
            heapq.heappush(hq, m)
        else:
            pop = heapq.heappop(hq)
            print(pop)
```
