```python
t = int(input())

def fibonacci(n):
    Sum = 0
    if n == 1 or n == 2:
        Sum = 1
    else:
        Sum = fibonacci(n-2)+fibonacci(n-1)
    return Sum
    
for _ in range(t):
    n = int(input())
    print(fibonacci(n))
```

```python
def Fibo(n):
    if n <= 2:
        return 1
    return Fibo(n-1) + Fibo(n-2)

T = int(input())

for _ in range(T):
    n = int(input())
    print(Fibo(n))
```

```python
def Fibo(n):
    if n == 0:
        return 0
    else:
        f=[0,1]
        for i in range(2,n+1):
            f.append(f[i-1]+f[i-2])
    return f[n]

T = int(input())

for _ in range(T):
    n = int(input())
    print(Fibo(n))
```

```python
def Fibo(n):
    if n <= 2:
        return 1
    x, y = 0, 1
    for _ in range(2, n+1):
        S = x + y
        x = y
        y = S
    return S
        
T = int(input())

for _ in range(T):
    n = int(input())
    print(Fibo(n))
```
