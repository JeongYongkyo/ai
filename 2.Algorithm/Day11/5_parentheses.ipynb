```python
def parentheses(text):
    match = {
        ')': '(',
        '}': '{',
        ']': '['
    }
    L = []
    for i in text:
        if i in '({[':
            L.append(i)
        elif i in match:
            if len(L) == 0:
                return("NO")
            else:
                if L.pop() != match[i]:
                    return("NO")
    if len(L) == 0:
        return("YES")
    else:
        return("NO")

t = int(input())

for _ in range(t):
    p = input()
    print(parentheses(p))
```

```python
t=int(input())

for _ in range(t):
    bracket=input()
    s=[]
    correct = True
    for b in bracket:
        if b == "(" or b == "{" or b == "[":
            s.append(b)
        elif b == ")":
            if not s or s.pop() != "(":
                correct=False
                break
        elif b == "}":
            if not s or s.pop() != "{":
                correct=False
                break
        elif b == "]":
            if not s or s.pop() != "[":
                correct=False
                break
    if s:
        correct=False
    if correct:
        print("YES")
    else:
        print("NO")
```
