````python
import torch
import torch.nn
````

````python
vocab = {
    'POSCO': 0,
    'AI': 1,
    '전문가': 2,
    'NLP': 3,
    '인공지능': 4,
    'embedding': 5,
    '실습': 6
}
````

````python
dim=10
emb_mtx = torch.nn.Embedding(len(vocab), dim)
````

````python
print(emb_mtx) # 7개단어를 10차원으로 표현
````

````python
print(emb_mtx.weight) # 초기의 random initial vector
````

# Problem 1
Print the word embedding of 'POSCO'

````python
idx = torch.tensor(vocab['POSCO'], dtype=torch.long) # make an index tensor for 'POSCO' (hint: the tensor's dtype is torch.long)
# feed the index tensor to emb_mtx
emb = emb_mtx(idx)
````

````python
print(idx, idx.size())
````

````python
print(emb, emb.size())
````

# Problem 2
Print the embeddings of the words in the sentence "POSCO AI 전문가"

````python
sent = ['POSCO', 'AI', '전문가']
idxs = []

for word in sent:
    idx = vocab[word] # convert a word into an index
    # append idx to idxs
    idxs.append(idx)

idxs = torch.tensor(idxs, dtype=torch.long) # convert the indices to a tensor
emb = emb_mtx(idxs) # matrix는 함수의 형태 # convert the indices into word embeddings
````

````python
print(idxs, idxs.size())
````

````python
print(emb, emb.size())
````

# Problem 3
Construct batch-level word embeddings for

```python
["POSCO AI 전문가", "NLP embedding 실습"]
```

````python
batch = [['POSCO', 'AI', '전문가'],
         ['NLP', 'embedding', '실습']]
batch_idxs = []

for sent in batch:
    idxs = []
    for word in sent:
        idx = vocab[word] # convert a word into an index
        # append idx to idxs
        idxs.append(idx)
        # append idxs to batch_idxs
    batch_idxs.append(idxs)

batch_idxs = torch.tensor(batch_idxs, dtype=torch.long) # convert batch_idxs into a tensor
emb = emb_mtx(batch_idxs) # convert idxs into word embedding
````

````python
print(batch_idxs, batch_idxs.size())
````

````python
print(emb, emb.size())
````
