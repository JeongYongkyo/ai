```python
import torch
import torch.nn
import gensim
```

```python
vocab = {
#     'POSCO': 0,
    'AI': 1,
    '전문가': 2,
    '북한': 3,
    '인공지능': 4,
    '대한민국': 5,
    '실습': 6
}
```

```python
model = gensim.models.Word2Vec.load('./save_files/newskor.model')
```

```python
## vocab에 기재된 단어의, 학습된 임베딩 (weights) 출력 ##
model.wv['AI']
```

# Step 1
PyTorch Tensor로 변환하기

```python
# vocabulary 내 모든 단어의 weights를 PyTorch Tensor로 변환하는 함수
def get_weights(vocab, embd):
    weights = []
    for word in vocab.keys():
        w = embd.wv[word]
        weights.append(w)
    return torch.tensor(weights)
```

```python
weights = get_weights(vocab, model)
```

```python
print(weights.shape)
```

# Step 2
임베딩 생성하기

```python
emb_mtx = torch.nn.Embedding.from_pretrained(weights)
```

```python
# print the word embedding of 'POSCO'
idx = torch.tensor([vocab['AI']], dtype=torch.long) # make index tensor for 'POSCO'
print(emb_mtx(idx))
```
