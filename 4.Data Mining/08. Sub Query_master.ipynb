```python
%%sql
select *
from orders
where shipvia = 
    (select shipvia
    from orders
    group by shipvia
    order by count(*) desc
    limit 1)
```

Orders table   Top3     .

```python
%%sql
select CustomerID
from orders
group by CustomerID
order by count(*) desc
limit 3
```

```python
%%sql
select *
from orders
where customerid in 
    (select CustomerID
    from orders
    group by CustomerID
    order by count(*) desc
    limit 3)
order by customerid
```

    ,     UnitPrice   UnitPrice     .

```python
%%sql
select categories.categoryname
from products
inner join categories
on products.categoryid = categories.categoryid
group by products.categoryid
having avg(products.unitprice) > (select avg(unitprice) from products)
```

```python
%%sql
select categories.categoryname
from (
    select categoryid, avg(unitprice) as 'avgprice'
    from products
    group by categoryid) A
inner join categories
on A.categoryid = categories.categoryid
where A.avgprice > (select avg(unitprice) from products)
```
