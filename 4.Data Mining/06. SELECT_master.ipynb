Products table CategoryID, SupplierID    . (    groupby)

```python
%sql SELECT CategoryID, SupplierID, Count(*) as 'Count' \
FROM  Products \
GROUP BY CategoryID, SupplierID
```

Products table CategoryID, SupplierID    . , CategoryID, SupplierID 2   .

```python
%sql SELECT CategoryID, SupplierID, Count(*) FROM Products \
WHERE CategoryID >= 2 AND SupplierID >= 2 \
GROUP BY CategoryID, SupplierID
```

Products table CategoryID, SupplierID    . ,   2  .

```python
%sql SELECT CategoryID, SupplierID, Count(*) AS 'Number' FROM Products \
GROUP BY CategoryID, SupplierID \
Having CategoryID >= 2 AND SupplierID >= 2
```

```python
%sql SELECT CategoryID, SupplierID, Count(*) AS 'Count' FROM Products \
GROUP BY CategoryID, SupplierID \
Having Count(*) >= 2 \
ORDER BY Count ASC
```
