## EDA : 탐색적 데이터 분석


- 다양한 기법(그래프)을 활용, 데이터에 대한 인사이트 극대화
- 다양한 관점에서 데이터를 이해하고, 데이터 분석 방향 설정
- 데이터 구조, 중요 변수, 이상치(outlier) 파악 및 가설 검증 활용
- 가정에서 시작하는 것이 아니라, 데이터로부터 연관 관계를 파악하고 적합한 모델링 방향 설정
- 탐색적 분석 방법: 문제 -> 데이터 -> 분석 -> 모델 -> 결론
    - 전통적인 분석 방법: 문제 -> 데이터 -> 모델 -> 분석 -> 결론


## Plotting and Visualization


## 파이썬에서 EDA 방법

- Collections(List, Tuple), Numpy -> Matplotlib 사용 (import matplotlib.pyplot as plt)
- Pandas(Series/DataFrame) -> 자체 그래프 함수 사용 (import pandas as pd)

## Matplotlib
    - plot([x], y, [**kwawrgs, ...]) : 데이터 수가 많은 경우
    - bar(x, y, [width, log, ...]) : 데이터 수가 적은 경우
    - hist(x, [bins, density, ...]) : 데이터 수가 많은 경우 요약
    - boxplot(x, [...]) : 데이터의 분포 요약
    - scatter(x, y, [...]) : 두 데이터 사이의 관계

Pandas 패키지를 활용해 주어진 tips.csv 파일을 불러온 후, 표시하시오.

```python
# csv 읽어오기
tips = pd.read_csv('./data/tips.csv')
tip = tips['tip']
tips.head()
```

tips dataframe 내 total_bill, tip, size 데이터를 numpy ndarray형태로 저장하시오.

```python
tips_np = tips.values
total_bill = tips_np[:,0] #첫번째 컬럼 데이터
tip = tips_np[:,1] #두번째 컬럼 데이터
size = tips_np[:,6] # 일곱번째 컬럼 데이터
```

### plot()

total_bill list를 plot 함수를 사용하여 그려보시오.

```python
plt.plot(total_bill)
```

tip list를 plot 함수를 사용하여 그려보시오.
- (세부 조건 설정) graph 색상 : 초록색, graph 타입 : 점선, graph 두께 : 2

```python
plt.plot(tip,'g--', linewidth=2)
```

### bar()

tip list의 앞 30개의 데이터만 bar 함수를 사용하여 그려보시오.

```python
plt.bar(range(tip[:30].size), tip[:30])
```

### hist()

tip list를 hist 함수를 사용하여 그려보시오.
- (세부 조건 설정) bin : [0,2,4,6,8,10]로 설정, y축 밀도 함수로 표현

```python
plt.hist(tip, bins=range(0,12,2), density=True) #[0,2,4,6,8,10]
```

### boxplot()

total_bill list를 boxplot 함수를 사용하여 그려보시오.


```python
plt.boxplot(total_bill, labels=['total_bill'])
```

### scatter()

total_bill과 tip 사이의 관계를 scatter 함수를 사용하여 그려보시오.
- (세부 조건 설정) 마커의 크기 80으로 설정, 마커의 색상을 tip에 연관되게 설정

```python
plt.scatter(total_bill, tip, s=80, c=tip)
```

### 두 데이터의 관계 보기

total_bill과 tip을 한 그래프 안에 표현하시오. (plot 함수 사용)

```python
plt.plot(total_bill)
plt.plot(tip)
```

```python
# 뒤에 해당 내용 있는지 확인하고 있으면 빼기
plt.plot(total_bill, label="a")
plt.plot(tip, label='b')
plt.legend()
```

두 개의 boxplot을 한 그래프 안에 표현하시오.

```python
plt.boxplot([total_bill, tip], labels=["total_bill", "tips"])
```

- 4개의 subplot을 그려보시오.
    - a. total_bill, tip의 산점도
    - b. total_bill 선 그래프
    - c. tip 선 그래프 (색 변경)

```python
fig, axes = plt.subplots(2, 2)
axes[0,0].scatter(total_bill, tip)
axes[0,1].plot(total_bill)
axes[1,0].plot(tip,'g-')
axes[1,1].plot(total_bill)
axes[1,1].plot(tip,'g-')
```

- total_bill과 tip의 그래프에서 아래 조건에 맞게 여러 정보들을 표시하고 범위를 변경하시오.
    - 각 선 그래프가 어떤 정보에 대한 그래프인지 label 표시 (hint : label, legend)
    - x ,y축 range 및 눈금 설정
    - 그래프 제목, x, y축 이름 설정

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

# 1) 
ax.plot(tip, label='tip')
ax.plot(total_bill, label='total_bill')
plt.legend(loc='best')

# 2)
ax.set_xlim([-50,260])
ax.set_ylim([-10,55])
ax.set_xticks([-50, 0, 100, 200])
ax.set_yticks(range(-10,60,5))

# 3) 
ax.set_title('My first matplotlib plot')
ax.set_xlabel('Customer')
ax.set_ylabel('USD')
```

```python
fig.savefig('figpath.svg') # 저장
```

## Plotting functions in pandas
    - plot() : 데이터 수가 많은 경우
    - plot(kind='bar') : 데이터 수가 적은 경우
    - plot(kind='hist') : 데이터 수가 많은 경우 요약
    - plot(kind='kde') : 데이터 분포 추정
    - plot(kind='box') : 데이터의 분포 요약
    - scatter(x, y, [...]) : 두 데이터 사이의 관계
