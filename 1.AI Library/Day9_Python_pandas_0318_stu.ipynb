# 7. 새로운 열 추가 및 수정
* 열이름 **total** 추가

```python
df['total'] = 0
df
```

```python
# total =  mid + final
df['total'] = df['mid'] + df['final']
df
```

* **실습)** 열이름 average 추가

```python
# 실습: 평균 = (Mid + Final) / 2 
df['average'] =(df['mid'] + df['final']) / 2
#df['average'] = df['total'] / 2
df
```

 
* **실습**) grade 추가 60점이상이면 pass 아니면 fail

```python
df['grade'] = df['average'] >= 60
df
```

```python
df[ df['average'] >= 60 ]
```

* **실습**) grade 항목 삭제 : del df['열이름']

```python
del df['grade']
df
```

* **열이름 Grade 추가**
    * apply() 사용 

```python
def calc_grade(tot):
    if tot >=90:
        grade = 'A'
    elif tot >=80:
        grade = 'B'
    elif tot >=70:
        grade = 'C'
    elif tot >=60:
        grade = 'D'
    else:
        grade = 'F'
        
    return grade        
```

```python
df['grade'] = df['average'].apply(calc_grade)
df
```
