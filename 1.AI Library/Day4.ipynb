### 파일 쓰기
"w"모드는 파일을 열어서 맨 앞을 가리킴 (기존 파일이 있으면 덮어쓰기됨)

```python
fw = open("new.txt","w")

for i in range(1,6):
    data = "%d line\n" % i
    fw.write(data)
        
fw.close()
```

### 파일 읽기
"r"모드로 열면 파일을 열어서 맨 앞을 가리킴

```python
fr = open("new.txt","r")

while True:
    line = fr.readline() #맨 앞줄부터 한줄 씩 읽어나가고 마지막에는 빈문자열 반환
    if line == "":
        break
    print(line, end="")

fr.close()
```

### 실습1
파일에 있는 각각의 단어의 수 구하기

```python
f = open("test.txt","r")

word_dic = dict()

data = f.read().split()

for line in data:
    if line in word_dic:
        word_dic[line] += 1
    else :
        word_dic[line] = 1

#print(word_dic)
for k, v in word_dic.items():
    print(k, v)
f.close()
```

### len()이 없다면?

```python
a = [5,5,6,7,8,3]
b = 'Life is short.'
c = [1,2,3]

length_a = 0
for i in a:
    length_a += 1
    if i == 0:
        break
print(length_a)

length_b = 0
for i in b:
    length_b += 1
    if i == 0:
        break
print(length_b)

length_c = 0
for i in c:
    length_c += 1
print(length_c)
```

```python
def my_len(l):
    length = 0
    for i in l:
        length += 1
    return length

a = [5,5,6,7,8,3]
b = 'Life is short.'
c = [1,2,3]

print(my_len(a), len(a))
print(my_len(b), len(b))
print(my_len(c), len(c))
```

### 실습문제 2
양의 정수 n을 전달받아 1부터 n까지 곱해주는 factorial 함수를 작성하시오

```python
def factorial(n):
    result = 1
    for i in range(n,1,-1):
        result *= i
    return result

factorial(3)
```

### 실습문제3
n과 a를 전달받아 를 계산하는 거듭제곱 함수를 작성하시오.

```python
def pow(n, a):
    result = n**a
    return result

pow(5, 3)
```

## 실습문제4
초를 전달받아 시간, 분, 초를 반환하는 함수를 작성하시오.

```python
def hour_min_sec(second):
    if(type(second) != int):
        print("정수를 입력바랍니다.")
    else:
        min = second / 60
        #print(min)
        hour = min / 60
        #print(hour)
        sec = second % 60
        #print(sec)
        min = min % 60
        #print(min)
    return hour, min, sec

hour, min, sec = hour_min_sec(57894)
print("%d시간 %d분 %d초" % (hour, min, sec))
```
