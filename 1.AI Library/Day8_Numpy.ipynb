```python
def hour_min_sec(s):
    h = s//3600
    m = (s%3600)//60
    s = (s%3600)%60
    return h,m,s

h,m,s = hour_min_sec(10007)
print(h,m,s)
```

```python
data = np.random.random((4,6)).round(1)
print(data, end='\n\n')
length = np.array([4,1,6,3])
print(length.shape, end='\n\n')
length = np.expand_dims(length, 1)
print(length.shape)
```

```python
maxlen = data.shape[1]
idx = np.arange(maxlen)
idx = np.tile(idx, (4,1))
print(idx, end='\n\n')
mask = idx < length
print(mask, end='\n\n')
data[~mask] = float('inf')
print(data)
```
