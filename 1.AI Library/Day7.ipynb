class DerivedClassName(BaseClassName):
    <statement-1>
    .

    .

```python
class A:
    a = 1
    def print_a():
        print("A class")

class B:
    b = 1
    def print_b():
        print("B class")
        
class C(A):
    c = 1
    def print_c():
        print("C class")

class D(C,B):
    d = 1
    def print_d():
        print("D class")
        
c = C()
d = D()

D.mro() #Method Resolution Order
```

### 실습 문제1
은행 계좌 클래스 만들기

```python
class BankAccount:
    def __init__(self, name='none', balance=0):
        self.name = name
        self.balance = balance
    
    def deposit(self, amount):
        self.balance += int(amount)
    
    def withdraw(self, amount):
        if self.balance < int(amount):
            print("잔액부족")
        else:
            self.balance -= int(amount)
    
    def get_info(self):
        print("이름 : ",self.name)
        print("잔고 : ",self.balance)    
```

```python
a = BankAccount("Gildong Hong", 100)
```

```python
a.deposit(400)
a.get_info()
```

```python
a.withdraw(600)
```

### 실습 문제2

```python
class MinimumBalanceAccount(BankAccount):
    def __init__(self, balance=0, name="none", min_bal=0):
        self.name = name
        self.balance = balance
        self.min_bal = min_bal
    
    def withdraw(self, amount):
        if self.balance - amount < self.min_bal:
            print("최소 잔액을 유지해야 합니다.")
            print("현재 잔액은",self.balance,"이고 최소 잔액은",self.min_bal,"입니다.")
        else:
            super().withdraw(amount)
```

```python
a = MinimumBalanceAccount(500, "Kim",100)
a.deposit(1000)
a.withdraw(1400)
```

```python
a.get_info()
```

### 과제 (1page)
Python의 기본 자료구조 set를 모방한 클래스 Set을 구현하세요

```python
class Set:
    def __init__(self, *val):
        self.setlist = []
        for x in val[0]:
            if x not in self.setlist:
                self.setlist.append(x)
        
    def add(self, elem):
        if elem not in self.setlist:
            self.setlist.append(elem)
        return self.setlist
    
    def discard(self, elem):
        if elem in self.setlist:
            self.setlist.remove(elem)
        return self.setlist
    
    def clear(self):
        self.setlist.clear()
        return self.setlist

    def __len__(self):
        return len(self.setlist)
    
    def __str__(self):
        self.setStr = "{"
        for x in self.setlist:
            self.setStr += str(x)
            self.setStr += ", "
        self.setStr = self.setStr[:-2]
        self.setStr += "}"
        return '%s'%(self.setStr)
    
    def __contains__(self, elem):
        if str(elem) in self.setStr:
            return True
        else:
            return False

    def __le__(self, elem, other):
        print(self.setStr)
        print(elem)
        print(other)
        
        
```

```python
a = Set([1,2,3,4])
b = Set([1,2,3,4])

print(a)
print(b)
print()

a.discard(4)
b.discard(1)
print(a)
print(b)
print()

print(len(a))
print(1 in a)
print(1 in b)
print()

print(a <= b)
```
