### break, continue

### ★출력 프로그램

```python
star = input("input an integer :")
for x in star :
    for i in range(int(x)) :
        print("★", end='')
    print("")
```

```python
star = input("input an integer :")
cnt = 0
while len(star) > cnt:
    for i in range(int(star[cnt])) :
        print("★", end='')
    cnt += 1
    print("")
```

### 실습1
“done”을 입력할 때까지 사용자로부터 숫자를 입력 받아 리스트에 저장하고, 
“done”을 입력하면, 리스트의 평균, 최대값과 최소값을 출력하는 프로그램을
작성하시오. (힌트. sum(), max(), min() 함수를 사용)

```python
my_num = input("Enter a number : ")
my_list = []

while True:
    if my_num != "done":
        my_list.append(my_num)
        my_num = input("Enter a number : ")
    else:
        print("")
        my_list = list(map(float, my_list))
        
        print(my_list)
        print("Average: %.1f" % (sum(my_list)/len(my_list)))
        print("Maximum: %.1f" % max(my_list))
        print("Minimum: %.1f" % min(my_list))        
        break
```
