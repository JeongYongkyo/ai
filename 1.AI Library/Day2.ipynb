### while문
1부터 10까지의 합(while 문 이용해서 작성)

```python
number = 1
total = 0
while number <= 10:
    total += number
    number += 1
print("1부터 10까지의 합:",total)
```

1부터 10까지의 곱(while 문 이용해서 작성)

```python
number = 1
prod = 1
while number <= 10:
    prod *= number
    number += 1
print("1부터 10까지의 곱:",prod)
```

```python
number = 1
while number <= 5:
    print(number, sep ="", end=", ")
    number += 1
```

### 반복문 - 실습1(구구단 출력)
입력한 숫자의 구구단을 출력하는 프로그램 작성하기

```python
gugudan = int(input("출력하고 싶은 단을 입력하세요:"))
for x in range(1, 10) :
    print(gugudan,"*",x,"=",gugudan*x)
```
