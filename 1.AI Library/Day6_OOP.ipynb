```python
def make_account():
    return {'balance' : 0}

def deposit(account, amount):
    account['balance'] += amount
    return account['balance']

def withdraw(account, amount):
    account['balance'] -= amount
    return account['balance']

a = make_account()
b = make_account()
print(a['balance'])
print(b['balance'])
print(deposit(a, 1000))
print(deposit(b, 2000))
print(withdraw(a, 500))
print(withdraw(b, 1000))
```

```python
# a() --> a.__call__() 특수 메서드

# __init__ 특수 메서드는 return은 None이어야함(생략해야함)
```

```python
class Person:
    def __init__(self, name, age, height, weight):
        self.name = name
        self.age = int(age)
        self.height = int(height)
        self.weight = int(weight)
    
    def weight_plus(self,value):
        self.weight += value
        return self.weight
    
    def height_plus(self,value):
        self.height += int(value)
        return self.height
        
    def get_info(self):
        print("1. 이름: ",self.name)
        print("2. 나이: ",self.age)
        print("3. 키: ",self.height)
        print("4. 몸무게: ",self.weight)
```

```python
kwon = Person('권홍석','25','170','65')
kwon.get_info()
print("===========================")
kwon.weight_plus(50)
kwon.get_info()
print("===========================")
kim = Person('김은희','26','160','55')
kim.get_info()
print("===========================")
kim.height_plus(10)
kim.get_info()
print("===========================")
youn = Person('윤은영','35','162','55')
youn.get_info()
```

```python
list_person = [kwon, kim, youn]
for p_obj in list_person:
    print(p_obj.name)
    print(p_obj.age)
    print("------------")
```

### 실습 문제1
은행 계좌 클래스 만들기

```python
class BankAccount:
    def __init__(self, name='none', balance=0):
        self.name = name
        self.balance = balance
    
    def deposit(self, amount):
        self.balance += int(amount)
    
    def withdraw(self, amount):
        if self.balance < int(amount):
            print("잔액부족")
        else:
            self.balance -= int(amount)
    
    def get_info(self):
        print("이름 : ",self.name)
        print("잔고 : ",self.balance)    
```

```python
a = BankAccount("Gildong Hong", 100)
```

```python
a.deposit(400)
a.get_info()
```

```python
a.withdraw(600)
```

```python
b = BankAccount("Kim", 0)
```

```python
b.deposit(1000)
b.get_info()
```

```python
b.withdraw(900)
b.get_info()
```
