## 실습문제4
초를 전달받아 시간, 분, 초를 반환하는 함수를 작성하시오.

```python
def hour_min_sec(second):
    if(type(second) != int):
        print("정수를 입력바랍니다.")
    else:
        m = second / 60
        h = m / 60
        s = second % 60
        m = m % 60
    return h, m, s

h, m, s = hour_min_sec(12312)
print("%d시간 %d분 %d초" % (h, m, s))
```

### 실습문제5
숫자로 구성된 리스트를 전달받아 최대값, 최소값을 반환하고, 두 수를 리스트에서 제거하는 함수를 작성하시오

```python
def get_min_max(l):
    a = min(l)
    b = max(l)
    l.remove(a)
    l.remove(b)
    return a,b
    
l = [3, 5, 9, 1, 2]
min_val, max_val = get_min_max(l)
print(min_val)
print(max_val)
print(l)
```

### 실습문제8
1. 전달받은 문장에서 각 알파벳이 몇 개씩 나오는지 저장하는 딕셔너리를 만든 후, 반환하는 함수 letter_dict()를 작성하시오.
2. letter_dict() 함수를 통해 생성된 딕셔너리에서 가장 많이 등장한 알파벳을 찾아 반환하는 함수를 작성하시오 (여러 개일 경우 아무 값이나 하나만 반환)

```python
def letter_dict(str):
    dic = dict()
    for i in str:
        if i not in dic:
            dic[i] = 1
        else:
            dic[i] += 1
    return dic

def max_letter(dict):
    new_dict = sorted(dict, key= lambda x : dict[x], reverse = True)
    return new_dict[0]

a = letter_dict('red apple')
print(a)
print(max_letter(a))
```

### 실습문제9
1. letter_dict() 함수에 'yellow banana＇를 전달하여 새로운 딕셔너리 b를 생성한다.
2. 두 딕셔너리 a, b를 전달받아 각 딕셔너리에 등장한 값을 모두 더한 새로운 딕셔너리 c를 반환하는 함수 comb_dict()를 작성하시오.

```python
def comb_dict(dict1, dict2):
    dic = dict1
    for k,v in dict2.items():
        if k not in dict1:
            dic[k] = 1
        else:
            dic[k] += v
    return dic

b = letter_dict('yellow banana')
c = comb_dict(a, b)
print(c)
print(max_letter(c))
```

### 과제1 prime.py
소수(Prime Number) 판단하는 함수 정의하기

```python
def check_prime(num):
    cnt = 0
    for i in range(1,int(num)+1):
        if num % i == 0:
            cnt += 1
    if cnt <= 2:
        return True
    else:
        return False
            
def main():
    a = 13
    b = 15
    
    if check_prime(a):
        print(str(a)+'는 소수입니다.')
    else:
        print(str(a)+'는 소수가 아닙니다.')
    
    if check_prime(b):
        print(str(b)+'는 소수입니다.')
    else:
        print(str(b)+'는 소수가 아닙니다.')
        
if __name__ == '__main__':
    main()
```

### 과제 3 - tok.py
전달받은 단어열의 n-gram을 구해주는 tokenize() 함수 정의하기

```python
def tokenize(trg, N=1):
    trg = trg.split()                 
    ngramlist = []
    for i in range(len(trg) - N+1):
        word1 = trg[i:i+N]
        word2 = (' ').join(word1)
        ngramlist.append(word2)
    return ngramlist
    
def main():
    a="There was a farmer who had a dog ."
    print(tokenize(a))
    print(tokenize(a, 2))
    
if __name__ == '__main__':
    main()
```

### 과제 4 - mean_and_var.py
여러 벡터를 전달받아, 평균(Mean)과 분산 (Variance)을 구하는 함수 정의하기

```python
def mean_and_var(*val):
    num, sum1, sum2, var1, var2 = 0, 0, 0, 0, 0
    for i in val:
        num += 1
        sum1 += float(i[0])
        sum2 += float(i[1])    
    m = sum1/num, sum2/num
    
    for x in val:
        var1 += (float(x[0] - m[0]))**2
        var2 += (float(x[1] - m[1]))**2
    v = var1/num, var2/num
        
    return m, v
        
def main():
    v1=(0, 1)
    v2=(0.5, 0.5)
    v3=(1, 0)
    
    m, var = mean_and_var(v1, v2, v3)
    print('평균: ', m)
    print('분산: ', var)

if __name__ == '__main__':
    main()
```
