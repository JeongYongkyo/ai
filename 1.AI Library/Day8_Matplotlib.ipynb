```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.linspace(0,10,100) #0부터 10사이의 값을 100개
b = np.exp(-a)
plt.plot(a,b,'*--g')  #x좌표, y좌표, String('Markers LineStyles Colors') 
plt.show()
```

```python
from numpy.random import normal

x = normal(size=200) # size 200개
plt.hist(x, bins=30) # 정규 분포안에 들어있는 값들을 30개의 통에 나눠 담는 결과를 히스토그램으로 표현
plt.show()
```

```python
from numpy.random import rand

a = rand(100) # random한 값 100개를 생성
b = rand(100)
plt.scatter(a,b)
plt.title("Scatter")
plt.show()
```

```python
x = np.linspace(0,2,100)

plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')

plt.xlabel('x label')
plt.ylabel('y label')

plt.title("Simplt plot")

plt.legend(loc='center')

plt.show()
```

```python
x = np.random.randn(100000)
y = .4 * np.random.randn(100000) +5

fig, axs = plt.subplots(1,2, sharey= True)

axs[0].hist(x, bins=20)
axs[1].hist(y, bins=20)

plt.show()
```

```python
x = np.linspace(0.01, 3.0, 200)

fig, axs = plt.subplots(2,2, sharex=True, figsize=(9,9))

axs[0,0].plot(x, x, '-g' ,label='linear')
axs[0,0].legend()
axs[0,0].set(xlabel='x', ylabel='y')
axs[0,1].plot(x, x**2, '--r', label='quadratic')
axs[0,1].legend()
axs[1,0].plot(x, np.exp(x), ':b', label='exponential')
axs[1,0].legend()
axs[1,1].plot(x, np.log(x), '-.y',label='logarithmic')
axs[1,1].legend(loc=4)

fig.text(0.5, 0.04, 'common x', ha='center')
fig.text(0.04, 0.5, 'common y', ha='center', rotation='vertical')
fig.suptitle("Simplt plot", y=0.95)

plt.show()
```
