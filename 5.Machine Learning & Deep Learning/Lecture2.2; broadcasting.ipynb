M1([m, n])&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;V1:[n] -> V1'[n,1]&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;T1:[c,m,n]  
M2([n, r])&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;V2:[n] -> V2'[n,1]&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;T2:[c,n,r]  

** 수학적  
&nbsp;&nbsp;M1M2&nbsp;&nbsp;&nbsp;M1V1  
&nbsp;&nbsp;[m,r]&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;[m,1]  

** pytorch  
1&nbsp;&nbsp;V1 @ V2 (내적곱)  
2&nbsp;&nbsp;M1 @ V1 (M1 @ V1' -> squeeze)  
3&nbsp;&nbsp;T1 @ V1  -> [T1 @ V1][i, :]  
4&nbsp;&nbsp;T1 @ T2 (tensorflow에서도 가능)
