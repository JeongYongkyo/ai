```python
df = pd.read_json("./data_files/reviews_Digital_Music_5.json", lines=True)
df.head()
```

```python
df.shape
```

```python
review_text = df['reviewText'].apply(gensim.utils.simple_preprocess)
```

```python
review_text
```

```python
df.reviewText.loc[0]
```

```python
model = gensim.models.Word2Vec(window=10, min_count=2, workers=4) # min_count : 총 빈도가 이보다 낮은 모든 단어를 무시
```

```python
model.build_vocab(review_text, progress_per=1000)
```

```python
model.train(review_text, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
model.save("./data_files/word2vec/word2vec.model")
```

```python
model.wv.similarity(w1='happy', w2='believe')
```

```python
model.wv.similarity(w1='happy', w2='smile')
```
