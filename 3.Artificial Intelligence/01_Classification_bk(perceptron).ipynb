# Perceptron

<br>
$$g(x) = \omega_0 + \omega^Tx  = \omega_0 + \omega_1x_1 + \omega_2x_2  = 0$$

<br><br>
$$ 
\begin{align*}
\omega &= \begin{bmatrix} \omega_0 \\ \omega_1 \\ \omega_2\end{bmatrix}\\ \\
x &= \begin{bmatrix} \left(x^{(1)}\right)^T \\ \left(x^{(2)}\right)^T \\ \left(x^{(3)}\right)^T\\ \vdots \\ \left(x^{(m)}\right)^T \end{bmatrix} = \begin{bmatrix} 1 & x_1^{(1)} & x_2^{(1)} \\ 1 & x_1^{(2)} & x_2^{(2)} \\ 1 & x_1^{(3)} & x_2^{(3)}\\\vdots & \vdots & \vdots \\ 1 & x_1^{(m)} & x_2^{(m)}\end{bmatrix}, \qquad
y = \begin{bmatrix}y^{(1)} \\ y^{(2)} \\ y^{(3)}\\ \vdots \\ y^{(m)} \end{bmatrix}
\end{align*}$$



## Perceptron Algorithm in Python

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#training data gerneration
m = 100
x1 = 8*np.random.rand(m, 1)
x2 = 7*np.random.rand(m, 1) - 4

g = 0.8*x1 + x2 - 3
```

```python
C1 = np.where(g >= 1)
C0 = np.where(g < -1)
print(C1)
```

```python
print(g.shape)
```

```python
C1 = np.where(g >= 1)[0]
C0 = np.where(g < -1)[0]
print(C1.shape)
print(C0.shape)
```

```python
x1[C1]
```

```python
x2[C1]
```

```python
# plot

plt.figure(figsize=(10,6))
plt.plot(x1[C1], x2[C1], 'ro', alpha=0.4, label='C1') # C1 label Data 필요
plt.plot(x1[C0], x2[C0], 'bo', alpha=0.4, label='C0') # C0 label Data 필요
plt.title('Linealry Separable Classes', fontsize = 15)
plt.legend(loc = 1, fontsize = 15)
plt.xlabel(r'$x_1$', fontsize = 15)
plt.ylabel(r'$x_2$', fontsize = 15)
plt.show()
```

$$ 
\begin{align*}
x &= \begin{bmatrix} \left(x^{(1)}\right)^T \\ \left(x^{(2)}\right)^T \\ \left(x^{(3)}\right)^T\\ \vdots \\ \left(x^{(m)}\right)^T \end{bmatrix} = \begin{bmatrix} 1 & x_1^{(1)} & x_2^{(1)} \\ 1 & x_1^{(2)} & x_2^{(2)} \\ 1 & x_1^{(3)} & x_2^{(3)}\\\vdots & \vdots & \vdots \\ 1 & x_1^{(m)} & x_2^{(m)}\end{bmatrix} \qquad 
y = \begin{bmatrix}y^{(1)} \\ y^{(2)} \\ y^{(3)}\\ \vdots \\ y^{(m)} \end{bmatrix}
\end{align*}$$

```python
#data X, y
X1 = np.hstack([np.ones([C1.shape[0],1]), x1[C1], x2[C1]])
X0 = np.hstack([np.ones([C0.shape[0],1]), x1[C0], x2[C0]])
X = np.vstack([X1, X0])

y = np.vstack([np.ones([C1.shape[0],1]), -np.ones([C0.shape[0],1])])

X = np.asmatrix(X)
y = np.asmatrix(y)
```

$$\omega = \begin{bmatrix} \omega_0 \\ \omega_1 \\ \omega_2\end{bmatrix}$$

<br>
$$\omega \leftarrow \omega + yx$$ where $(x, y)$ is a misclassified training point

```python
# perceptron

w = np.ones([3,1])
w = np.asmatrix(w)

n_iter = y.shape[0]
for _ in range(n_iter):
    for i in range(n_iter):
        if y[i,0] != np.sign(X[i,:]*w)[0,0]:
            w += y[i,0]*X[i,:].T
print(w)
```

$$
\begin{align*}
g(x) &= \omega_0 + \omega^Tx  = \omega_0 + \omega_1x_1 + \omega_2x_2  = 0 \\\\
\implies x_2 &= -\frac{\omega_1}{\omega_2} x_1 - \frac{\omega_0}{\omega_2}
\end{align*}
$$

```python
'''
plot

'''
x1p = np.linspace(0,8,100).reshape([-1,1])
x2p = - w[1,0]/w[2,0]*x1p - w[0,0]/w[2,0]

plt.figure(figsize=(10,6))
plt.plot(x1[C1], x2[C1], 'ro', alpha=0.4, label='C1') # C1 label Data 필요
plt.plot(x1[C0], x2[C0], 'bo', alpha=0.4, label='C0') # C0 label Data 필요
plt.plot(x1p, x2p, c='k', linewidth= 3, label='perceptron')
#plt.plot(x1p, x2p-1, '--k')
#plt.plot(x1p, x2p+1, '--k')
plt.title('Linealry Separable Classes', fontsize = 15)
plt.legend(loc = 1, fontsize = 15)
plt.xlabel(r'$x_1$', fontsize = 15)
plt.ylabel(r'$x_2$', fontsize = 15)
plt.show()
```

```python
'''
data X, y
(Hint : hstack, vstack, np.ones)
'''
X1 = np.hstack([x1[C1], x2[C1]])
X0 = np.hstack([x1[C0], x2[C0]])
X = np.vstack([X1, X0])

y = np.vstack([np.ones([C1.shape[0],1]), -np.ones([C0.shape[0],1])])
```

```python
from sklearn import linear_model

clf = linear_model.Perceptron(tol=1e-3)
clf.fit(X, np.ravel(y))
```

```python
clf.predict([[3, -2]])
```

```python
clf.predict([[6, 2]])
```
