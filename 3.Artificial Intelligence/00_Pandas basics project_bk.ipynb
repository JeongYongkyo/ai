## NaN 비율 확인

```python
'''
각 열마다, NaN 수 구하기
(isna, values, sum 사용)
'''
per_nan = df_total.isna().sum() / len(df_total)*100
```

```python
'''
각 열마다, NaN 수 비율 구하기
(shape 사용)
'''
check_nan = df_total.isna().values
check_nan = check_nan.sum(axis=0)
per_nan = (check_nan/df_total.shape[0])*100
per_nan
```

```python
print('year column 에서 NaN 비율(%) :', per_nan[0])
print('name_id_code column 에서 NaN 비율(%) :', np.round(per_nan[1],2))
print('points column 에서 NaN 비율(%) :', per_nan[2])
print('penalty column 에서 NaN 비율(%) :', np.round(per_nan[3],2))
```

## 데이터 선별 후, 비중 확인

```python
# points가 30 점 이상, 90 점 이하 데이터 선별
selected_data = df_total[(df_total['points']>=30) & (df_total['points']<=90)]
```

```python
selected_data
```

```python
per_selected_data = (selected_data.shape[0]/df_total.shape[0])*100
per_selected_data
```

```python
'''
선별된 데이터가 전체 데이터에서 차지하는 비중 구하기
'''
per_selected_data = len(selected_data) / len(df_total)*100
```

```python
print('selected data가 전체 data에서 차지하는 비중(%) :', np.round(per_selected_data,2))
```
