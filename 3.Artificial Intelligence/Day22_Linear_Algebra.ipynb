# Linear Algebra

## Linear Equations

```python
import numpy as np
```

```python
A = np.array([[4,-5],
              [-2,3]])
b = np.array([[-13],
                [9]])
```

```python
print(A)
```

```python
print(b)
```

```python
x = np.linalg.inv(A).dot(b)
# x = np.linalg.inv(A) @ b
print(x)
```

```python
A = np.asmatrix(A)
B = np.asmatrix(b)

x = A.I*B
print(x)
```

## Vector-Vector Products

```python
x = np.array([[1],
              [1]]) # (2, 1) -> (1, 2)

y = np.array([[2],
              [3]])
```

```python
print(x.T.dot(y))
```

```python
x = np.asmatrix(x)
y = np.asmatrix(y)

print(x.T*y)
```

## Norms (strenth or distance in linear space)

```python
x = np.array([[4],
              [3]])

np.linalg.norm(x,2)
```

```python
np.linalg.norm(x,1)
```

## Orthogonal Projection onto a Subspace

```python
import numpy as np

A = np.matrix([[1,0],
               [0,1],
               [0,0]])
B = np.matrix([[2],
               [2],
               [2]])
```

```python
X_star = (A.T*A).I*A.T*B
```

```python
print(X_star)
```

```python
A * X_star
```
