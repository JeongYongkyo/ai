```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
'''
Data generate
x, y좌표 생성
'''

x = np.array([0.1, 0.4, 0.7, 1.2, 1.3, 1.7, 2.2, 2.8, 3.0, 4.0, 4.3, 4.4, 4.9]).reshape(-1, 1)
y = np.array([0.5, 0.9, 1.1, 1.5, 1.5, 2.0, 2.2, 2.8, 2.7, 3.0, 3.5, 3.7, 3.9]).reshape(-1, 1)

plt.figure(figsize=(10, 8)) # figure 사이즈 (가로,세로)
plt.plot(x, y, 'ko')
# matplotlib marker
# pytorch.org
plt.title('Data', fontsize = 15)
plt.xlabel('X', fontsize = 15)
plt.ylabel('Y', fontsize = 15)
plt.axis('equal') # 가로 세로 축 간격 맞춤
plt.grid(alpha = 0.3) # alpha : 투명도
plt.xlim([0, 5]) # x에 대한 범위 지정
plt.show() # 생략 가능
# plt.savefig(''./test.png, dpi = 300)  #plt 저장, dpi 화질
```

```python
np.zeros(shape = (x.shape[0],1))
```

```python
'''
min (A*theta -y)**2

theta에 대해 미분하면
theta = (A.T * A).I * A.T * y
'''

A = np.hstack([x**0, x**1])
print(A)
```

```python
'''
theta = np.linalg.inv(A.T @ A) @ A.T @ y
 2가지 방법 , 최소제곱법
'''

A = np.asmatrix(A)
theta = (A.T @ A).I @ A.T @ y
print('theta:\n', theta)
```

```python
'''
xp는 0에서 5까지 그래프에 plot하기 위해 생성된 data
xp에 theta값을 반영하여 구하고자하는 직선(yp)을 figure에 plot
'''

# to plot
plt.figure(figsize = (10, 8))
plt.title('Regression', fontsize = 15)
plt.xlabel('X', fontsize = 15)
plt.ylabel('Y', fontsize = 15)
plt.plot(x, y, 'ko', label = "data")

# to plot a straight line (fitted line)
xp = np.arange(0, 5, 0.01).reshape([-1, 1])
xp_ = np.hstack([xp**0, xp])

#yp = xp * theta[1, 0] + theta[0, 0]
yp = xp_ @ theta
yp.shape

plt.plot(xp, yp, 'r', linewidth = 2, label = "regression")
plt.legend(fontsize = 15)
plt.axis('equal')
plt.grid(alpha = 0.3)
plt.xlim([0, 5])
plt.show()
```

```python
'''
Gradient Descent를 이용한 방법
min (A*theta -y)**2

gradient f = 2*(A.T*A*theta - A.T*y)
'''

theta = np.random.randn(2, 1)
alpha = 0.001

for _ in range(1000):
    df = 2 * (A.T @ A @ theta - A.T @ y)
    theta = theta - alpha * df
    
print (theta)
```

## Use CVXPY Optimization

```python
import cvxpy as cvx

theta2 = cvx.Variable([2, 1])

obj = cvx.Minimize(cvx.norm(A @ theta2 - y, 2))
constraints = []
prob = cvx.Problem(obj, constraints)
prob.solve()

print('theta:\n', theta2.value)
```

```python
import cvxpy as cvx

theta1 = cvx.Variable([2, 1])

obj = cvx.Minimize(cvx.norm(A @ theta1 - y, 1))
constraints = []
prob = cvx.Problem(obj, constraints)
prob.solve()

print('theta:\n', theta1.value)
```

```python
# to plot data
plt.figure(figsize = (10, 8))
plt.title('$L_1$ and $L_2$ Regression', fontsize = 15)
plt.xlabel('X', fontsize = 15)
plt.ylabel('Y', fontsize = 15)
plt.plot(x, y, 'ko', label = 'data')

# to plot straight lines (fitted lines)
xp = np.arange(0, 5, 0.01).reshape(-1, 1)
yp1 = xp * theta1.value[1, 0] + theta1.value[0, 0]
#yp2 = xp * theta2.value[1, 0] + theta1.value[0, 0]
yp2 = xp * theta2.value[1, :] + theta1.value[0, :] # 슬라이싱도 가능

plt.plot(xp, yp1, 'b', linewidth=2, label = '$L_1$')
plt.plot(xp, yp2, 'r', linewidth=2, label = '$L_2$')
plt.legend(fontsize = 15)
plt.axis('equal')
plt.xlim([0, 5])
plt.grid(alpha = 0.3)
plt.show()
```

```python
'''
Outlier가 포함된 경우
'''

x = np.array([0.1, 0.4, 0.7, 1.2, 1.3, 1.7, 2.2, 2.8, 3.0, 4.0, 4.3, 4.4, 4.9]).reshape(-1, 1)
y = np.array([0.5, 0.9, 1.1, 1.5, 1.5, 2.0, 2.2, 2.8, 2.7, 3.0, 3.5, 3.7, 3.9]).reshape(-1, 1)

# add outliers
x = np.vstack([x, np.array([0.5, 3.8]).reshape(-1, 1)])
y = np.vstack([y, np.array([3.9, 0.3]).reshape(-1, 1)])

A = np.hstack([x**0, x])
A.shape
```

```python
plt.figure(figsize = (10, 8))
plt.plot(x, y, 'ko', label = 'data')
plt.axis('equal')
plt.xlim([0, 5])
plt.grid(alpha = 0.3)
plt.show()
```

```python
theta2 = cvx.Variable([2,1])
obj2 = cvx.Minimize(cvx.norm(A @ theta2 - y, 2))
prob2 = cvx.Problem(obj2, [])
prob2.solve()

print('theta:\n', theta2.value)
```

```python
# to plot straight lines (fitted lines)
plt.figure(figsize = (10, 8))
plt.plot(x, y, 'ko', label = 'data')
xp = np.arange(0, 5, 0.01).reshape(-1,1)
yp2 = theta2.value[0,0] + theta2.value[1,0]*xp

plt.plot(xp, yp2, 'r', linewidth = 2, label = '$L_2$')
plt.axis('equal')
plt.xlim([0, 5])
plt.legend(fontsize = 15, loc = 5)
plt.grid(alpha = 0.3)
plt.show()
```

```python
theta1 = cvx.Variable([2,1])
obj1 = cvx.Minimize(cvx.norm(A @ theta1 - y, 1))
prob1 = cvx.Problem(obj1, [])
prob1.solve()

print('theta:\n', theta1.value)
```

```python
# to plot straight lines (fitted lines)
plt.figure(figsize = (10, 8))
plt.plot(x, y, 'ko', label = 'data')
xp = np.arange(0, 5, 0.01).reshape(-1,1)
yp1 = theta1.value[0,0] + theta1.value[1,0]*xp

plt.plot(xp, yp1, 'b', linewidth = 2, label = '$L_1$')
plt.axis('equal')
plt.xlim([0, 5])
plt.legend(fontsize = 15, loc = 5)
plt.grid(alpha = 0.3)
plt.show()
```

```python
# to plot data
plt.figure(figsize = (10, 8))
plt.plot(x, y, 'ko', label = 'data')
plt.title('$L_1$ and $L_2$ Regression w/ Outliers', fontsize = 15)
plt.xlabel('X', fontsize = 15)
plt.ylabel('Y', fontsize = 15)

# to plot straight lines (fitted lines)
xp = np.arange(0, 5, 0.01).reshape(-1,1)
yp1 = theta1.value[0,0] + theta1.value[1,0]*xp
yp2 = theta2.value[0,0] + theta2.value[1,0]*xp

plt.plot(xp, yp1, 'b', linewidth = 2, label = '$L_1$')
plt.plot(xp, yp2, 'r', linewidth = 2, label = '$L_2$')
plt.axis('scaled')
plt.xlim([0, 5])
plt.legend(fontsize = 15, loc = 5)
plt.grid(alpha = 0.3)
plt.show()
```
